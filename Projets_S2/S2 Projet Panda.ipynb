```python
import pandas as pd
hp = pd.read_csv('./Data/house_price.csv')
chb = pd.read_csv('./Data/nombre_de_chambres.csv')
sdb = pd.read_csv('./Data/nombre_de_sdb.csv')
sf = pd.read_csv('./Data/superficie.csv')
print('dataset loaded')
```

```python
chb.columns = ['ID', 'Nb chambres']
sdb.columns = ['ID', 'Nb Salles de Bain']
sf.columns = ['ID', 'Superficie']

datasetV1 = chb.join ( sdb.set_index( [ 'ID' ], verify_integrity=True ), on=[ 'ID' ], how='left' )
dataset = datasetV1.join ( sf.set_index( [ 'ID' ], verify_integrity=True ), on=[ 'ID' ], how='left' )
```
