```python
prix_caract = df.item_price.str[1:]
#prendre uniquement la valeur à partir du 2eme caractère
prix = df.item_price.str.replace('$', '').astype('float64')
df['item_price'] = prix
#retirer caractère $ et trasnformer en nombre à virgule
```
